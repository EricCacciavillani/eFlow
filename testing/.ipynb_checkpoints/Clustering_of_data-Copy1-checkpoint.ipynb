```python
def euclidean_distance(v1, v2):
    dist = [(a - b)**2 for a, b in zip(v1, v2)]
    dist = math.sqrt(sum(dist))
    return dist
```

```python
def find_all_dist_with_target(data_points,
                              target_point):
    """
        Finds all distances between the target and the other points.
    """
    distances = []
    for dp in data_points:
        distances.append(euclidean_distance(dp,target_point))
    
    return stats.zscore(np.array(distances))
```

```python
def multiprocess(func, jobs, cores):
    
    results = []
    
    if cores == 1:
        for j in jobs:
            results.append(func(j))
    elif cores == -1:
        with mp.Pool(mp.cpu_count()) as p:
            results = list(p.map(func, jobs))
    elif cores > 1:
        with mp.Pool(cores) as p:
            results = list(p.map(func, jobs))
    else:
        print('Error: cores must be a integer')
    
    return results
```
